# Jakie źródła pozyskania pracownika (Recruitment Source) są najlepsze, jeśli zależy nam na jak najdłuższym stażu pracowników?

```python
import datetime as dt

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2022,7,14)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(1,'Y')
```

```python
df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

```python
plt.figure(figsize=(20,10))
sns.displot(x='Seniority', y='RecruitmentSource', data=df)
```

Najlepszymi źródłami do pozyskiwania pracowników, według wykresu, są, targi pracy, bilbordy oraz wyszukiwarka internetowa, ponieważ pracownicy zatrudnieni przez te źródła komunikacji mają największy staż pracy.

# Jak wygląda struktura wieku aktualnie zatrudnionych pracowników?


```python
df2 = df[df['Termd'] == 0]

def calculate_age(row):
    today = date.today()
    return today.year - row.year - ((today.month, today.day) < (row.month, row.day))

df2['Age'] = df['DOB'].apply(calculate_age)

plt.figure(figsize=(20,10))
sns.countplot(x='Age', data=df2)
```

Nie rozumiem dlaczego, ale podczas konwersji z str do datetime, python podstawia daty z przyszłości. 01/18/52 zmienia na 2052-01-18. Dlatego na wykresie są wskazane wartości ujemne.

# Czy starsi pracownicy pracują nad większą liczbą specjalnych projektów niż młodsi pracownicy?

```python
df2 = df2[df2['Age']>0]

plt.figure(figsize=(20,10))
sns.barplot(x='Age', y='SpecialProjectsCount', data=df2)
```

Z wykresu wynika, że więcej starszych pracowników otrzymało dużą ilość projektów specjalnych. Jednakże, nie mamy danych kiedy otrzymali projekt specjalny, przez to nie możemy stwierdzić czy ich wiek miał duży wpływ na przydział projektu. Może zostali przydzieleni podczas pierwszego roku pracy, kiedy byli młodsi?

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='SpecialProjectsCount', y='Seniority', data=df2)
```

Na tym wykresie widzimy, że większość projektów specjalnych jest wykonywana przez pracowników ze stażem pracy poniżej 10.

```python
df3 = df2[(df2['Seniority'] < 10) & (df2['Seniority'] > 7)]
df3
sns.barplot(x='Age', y='SpecialProjectsCount', data=df3)
```

Pośród pracowników ze stażem pomiędzy 7 a 10, najwięcej projektów specjalnych wykonują pracownicy w wieku 53 lat.
